```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = '/content/HR Data.csv'
hr_data = pd.read_csv(file_path)
```

```python
# Step 1: Remove unnecessary columns
columns_to_remove = ['EmployeeCount', 'StandardHours', 'EmployeeNumber']
hr_data_cleaned = hr_data.drop(columns=columns_to_remove)
```

```python
# Step 2: Rename columns for clarity
columns_rename = {
    'Age': 'age',
    'Attrition': 'attrition',
    'BusinessTravel': 'business_travel',
    'DailyRate': 'daily_rate',
    'Department': 'department',
    'DistanceFromHome': 'distance_from_home',
    'Education': 'education',
    'EducationField': 'education_field',
    'EnvironmentSatisfaction': 'environment_satisfaction',
    'Gender': 'gender',
    'HourlyRate': 'hourly_rate',
    'JobInvolvement': 'job_involvement',
    'JobLevel': 'job_level',
    'JobRole': 'job_role',
    'JobSatisfaction': 'job_satisfaction',
    'MaritalStatus': 'marital_status',
    'MonthlyIncome': 'monthly_income',
    'MonthlyRate': 'monthly_rate',
    'NumCompaniesWorked': 'num_companies_worked',
    'Over18': 'over_18',
    'OverTime': 'over_time',
    'PercentSalaryHike': 'percent_salary_hike',
    'PerformanceRating': 'performance_rating',
    'RelationshipSatisfaction': 'relationship_satisfaction',
    'StockOptionLevel': 'stock_option_level',
    'TotalWorkingYears': 'total_working_years',
    'TrainingTimesLastYear': 'training_times_last_year',
    'WorkLifeBalance': 'work_life_balance',
    'YearsAtCompany': 'years_at_company',
    'YearsInCurrentRole': 'years_in_current_role',
    'YearsSinceLastPromotion': 'years_since_last_promotion',
    'YearsWithCurrManager': 'years_with_curr_manager'
}

hr_data_cleaned.rename(columns=columns_rename, inplace=True)
```

```python
# Step 3: Eliminate redundant entries
hr_data_cleaned = hr_data_cleaned.drop_duplicates()
```

```python
# Eliminate NaN values
hr_data_cleaned = hr_data_cleaned.dropna()
```

```python
# Save the cleaned data to a new CSV file
output_file_path = '/content/HR Data Cleaned.csv'
hr_data_cleaned.to_csv(output_file_path, index=False)
```

```python
# Display a summary and the first few rows of the cleaned data
print(hr_data_cleaned.info())
```

```python
print(hr_data_cleaned.head())
```

```python
# Summary statistics for numerical columns
summary_numerical = hr_data_cleaned.describe()
```

```python
# Summary statistics for categorical columns
summary_categorical = hr_data_cleaned.describe(include=['object'])
```

```python
# Attrition rate
attrition_rate = hr_data_cleaned['attrition'].value_counts(normalize=True) * 100
```

```python
# Attrition rate by department
attrition_by_department = hr_data_cleaned.groupby('department')['attrition'].value_counts(normalize=True).unstack() * 100
```

```python
# Select only the numeric columns for correlation matrix
numeric_columns = hr_data_cleaned.select_dtypes(include='number')
```

```python
# Visualization
# Distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(hr_data_cleaned['age'], kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Attrition by Department
plt.figure(figsize=(10, 6))
attrition_by_department.plot(kind='bar', stacked=True)
plt.title('Attrition Rate by Department')
plt.xlabel('Department')
plt.ylabel('Attrition Rate (%)')
plt.show()
```

```python
# Compute correlation matrix for numerical columns
correlation_matrix = numeric_columns.corr()
```

```python
# Heatmap of correlation matrix
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Display the computed summaries
print("Summary Statistics for Numerical Columns:")
print(summary_numerical)

print("\nSummary Statistics for Categorical Columns:")
print(summary_categorical)

print("\nAttrition Rate (%):")
print(attrition_rate)

print("\nAttrition Rate by Department (%):")
print(attrition_by_department)
```
